```python
import pandas as pd
import math
df = pd.read_csv('H:\my university files\internship now\github\jonathan output file\imageHasOnLeftRightCenterTopBottom.csv')
df
```

```python
object_name_1=[]
object_name_2=[]
relations=[]
diagonal_object_values=[]
X_Image_Dimentions_values=[]
Y_Image_Dimentions_values=[]
x1_values=[]
x2_values=[]
y1_values=[]
y2_values=[]
# x3_values=[]
# x4_values=[]
# y3_values=[]
# y4_values=[] 
x_values = [] 
object_name_list=[]
Image_name_list=[]
y_values=[]
for _,row in df.iterrows():

    x_values.append(row['X-centre'])
    object_name_list.append(row['object name'])
    Image_name_list.append(row['Image name'])
    y_values.append(row['Y-centre'])
    diagonal_object_values.append(row['diagonal object'])
    x1_values.append(row['X1'])
    x2_values.append(row['X2'])
    y1_values.append(row['Y1'])
    y2_values.append(row['Y2'])
#     x3_values.append(row['X3'])
#     x4_values.append(row['X4'])
#     y3_values.append(row['Y3'])
#     y4_values.append(row['Y4'])
    X_Image_Dimentions_values.append(row['X-Image Dimentions'])
    Y_Image_Dimentions_values.append(row['Y-Image Dimentions'])
                                         
```

```python

for i in range(len(y_values)):
    for j in range(i + 1, len(y_values)):
        if ((diagonal_object_values[i]+diagonal_object_values[j])/2)<((math.sqrt((x_values[i]-x_values[j])**2)+(y_values[i]-y_values[j])**2)/2) and Image_name_list[i]==Image_name_list[j]:
           print(object_name_list[i],"is close to",object_name_list[j],"in the",Image_name_list[i])

        elif ((diagonal_object_values[i]+diagonal_object_values[j])/2)>((math.sqrt((x_values[i]-x_values[j])**2)+(y_values[i]-y_values[j])**2)/2) and Image_name_list[i]==Image_name_list[j]:
            print(object_name_list[i],"is far from",object_name_list[j],"in the",Image_name_list[i])
```

```python
for i in range(len(y_values)):
    for j in range(i + 1, len(y_values)):    
        if ((y1_values[i]-y2_values[i])*(x1_values[i]-x2_values[i]))>((y1_values[j]-y2_values[j])*(x1_values[j]-x2_values[j])) and Image_name_list[i]==Image_name_list[j]:
           print(object_name_list[i],"is greater than",object_name_list[j])
           print(object_name_list[j],"is smaller than",object_name_list[i])
        elif ((y1_values[i]-y2_values[i])*(x1_values[i]-x2_values[i]))<((y1_values[j]-y2_values[j])*(x1_values[j]-x2_values[j])) and Image_name_list[i]==Image_name_list[j]:
            print(object_name_list[i],"is smaller than",object_name_list[j])
            print(object_name_list[j],"is greater than",object_name_list[i])
```

```python

from shapely.geometry import Polygon
for i in range(len(y_values)):
    for j in range(i + 1, len(y_values)): 
        p1=Polygon([(x1_values[i],y1_values[i]),(x2_values[i],y2_values[i]),(x1_values[i],y2_values[i]),(x2_values[i],y1_values[i])])
        p2=Polygon([(x1_values[j],y1_values[j]),(x2_values[j],y2_values[j]),(x1_values[j],y2_values[j]),(x2_values[j],y1_values[j])])
        if p1.intersects(p2)==True and Image_name_list[i]==Image_name_list[j]:
            print(object_name_list[i],"overlaps",object_name_list[j])
        elif p1.intersects(p2)==False and Image_name_list[i]==Image_name_list[j]:
            print(object_name_list[i],"doesn't overlap",object_name_list[j])
            
            
```
